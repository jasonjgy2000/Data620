```python
%matplotlib inline
import networkx as nx
import networkx.algorithms.bipartite as bi
import matplotlib.pyplot as plt
```

```python
G = nx.davis_southern_women_graph()
women = G.graph['top']
clubs = G.graph['bottom']
```

```python
# project bipartite graph onto women nodes
W = bipartite.projected_graph(G, women)
plt.figure(figsize=(15,10))
nx.draw_networkx(W,with_labels=True)
plt.show()
```

The graph is well connected with key women Flora Price and Katherina Rogers

```python
# performing centrality analysis on the women's graph
nx.betweenness_centrality(W)
```

```python
# computing the affiliation network of then women
womenNet = bi.weighted_projected_graph(W, women, ratio=False)
# calculating the weight (strength) of relationships
wWeights = [edata['weight'] for f,t,edata in womenNet.edges(data=True)]
plt.figure(figsize=(18,12))
nx.draw_networkx(womenNet,width=wWeights, edge_color=weights)
plt.show()
```

```python
# project bipartite graph onto women nodes
C = bipartite.projected_graph(G, clubs)
plt.figure(figsize=(15,10))
nx.draw_networkx(C,with_labels=True)
plt.show()
```

```python
# creating club network
clubNet = bi.weighted_projected_graph(C, clubs, ratio=False)
# calculating the weight (strength) of relationships
weights = [edata['weight'] for f,t,edata in clubNet.edges(data=True)]
plt.figure(figsize=(10,15))
nx.draw_networkx(clubNet,width=weights, edge_color=weights)
plt.show()
```
