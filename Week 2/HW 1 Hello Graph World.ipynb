```python
%matplotlib inline
import networkx as nx
import matplotlib.pyplot as plt

g= nx.Graph()
g.add_nodes_from(['Jane','Ike','Heather','Fernando','Grath','Carol','Diane','Ed','Beverly','Andre'])
g.add_edges_from([('Jane','Ike'),('Ike','Heather'),('Heather','Fernando')
                 ,('Heather','Grath'),('Fernando','Carol'),('Carol','Andre')
                  ,('Andre','Beverly'),('Beverly','Ed'),('Ed','Grath')
                  ,('Carol','Diane'),('Diane','Fernando'),('Diane','Grath')
                  ,('Diane','Ed'),('Diane','Beverly'),('Diane','Andre')
                  ,('Fernando','Andre'),('Beverly','Grath'),('Fernando','Grath')                  
                 ])
nx.draw(g,with_labels=True)
plt.show()
```
